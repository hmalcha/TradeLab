# Notebook for testing trading strategies.

Import TradeLab package

```python
from tradelab import *
```

Fetch hisorical stock data

```python
data_handler = YFinanceDataHandler(
    tickers=["AAPL", "MSFT"],
    start="2010-01-01",
    end="2010-01-15"
)

portfolio = Portfolio(deposit=1000)

strategy = BuyAndHold(
    ticker= "AAPL",
    quantity=1,
    start_time= "2010-01-06",
    end_time= "2010-01-10"
)

engine = TradingEngine(
    portfolio = portfolio,
    strategy = strategy,
    data_handler = data_handler
)

history = engine.run_backtest()
print(history)
```
